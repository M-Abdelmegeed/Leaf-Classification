## Understanding The Dataset

The dataset comprises a total of 1584 records, with 990 instances allocated for training in the `train.csv` file and the remaining 594 instances used for testing in the `test.csv` file.

### Attributes:

- **id**: Identifier for each record
- **margin1 - margin64**: Numerical features representing leaf margin characteristics
- **shape1 - shape64**: Numerical features representing leaf shape characteristics
- **texture1 - texture64**: Numerical features representing leaf texture characteristics

The distribution of records in the dataset is as follows:

- **Training Data (`train.csv`):** 990 records
- **Testing Data (`test.csv`):** 594 records

### Images:

The dataset includes a folder named **images** containing 1584 black and white JPG images of leaves. Each image corresponds to a record in the dataset and can be referenced using the respective **id**.

## Preprocessing

### 1. Removing Highly Correlated Attributes

The initial step in preprocessing involves the identification and removal of highly correlated attributes. This is crucial for mitigating multicollinearity, ensuring that the model isn't influenced by redundant information. Highly correlated attributes are detected through correlation analysis, and one feature from each correlated pair is dropped.

### 2. Checking for Missing Values and Duplicates

Following the removal of highly correlated attributes, the dataset undergoes a check for missing values and duplicates. This step ensures data integrity and consistency.

### 3. Standardizing the Data

Standardization becomes the next step to ensure that all numerical features are on a similar scale. This is achieved by subtracting the mean and dividing by the standard deviation of each feature.

### 4. Encoding Categorical Variables

Encoding is applied to convert them into a format suitable for machine learning algorithms.

## 3. Standardizing the Data
Before we standardize the data, lets just take a look at how the data looks without standardizing, to see the difference between data before standardization and data after standardization.

```python
df_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Select only the numerical columns for standardization
numerical_columns = df_train.select_dtypes(include=['float64', 'int64']).columns

scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test data
df_train[numerical_columns] = scaler.fit_transform(df_train[numerical_columns])
df_test[numerical_columns] = scaler.transform(df_test[numerical_columns])
```

```python
df_train.head()
```

As we can see, now we have negative values after we standardized the data with the **StandardScaler**. Before the standardization we had no negative values at all.

## 4. Encoding Species
The process of encoding the species attribute involves transforming categorical information into a numerical format that can be effectively utilized by machine learning models or deep neural networks as we are going to explore later on.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_train['species_encoded'] = label_encoder.fit_transform(df_train['species'])
```
